```python
import subprocess
def run_cmd(args):
    r = subprocess.run(args, capture_output=True, text=True, shell=True)
    print("STDOUT:")
    print(r.stdout)
    print("STDERR:")
    print(r.stderr)

run_cmd("git status")
```
